```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_approved_at'] = pd.to_datetime(orders['order_approved_at'])
orders['order_delivered_carrier_date'] = pd.to_datetime(orders['order_delivered_carrier_date'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])
```

```python
orders.info()
```

```python
orders['order_year'] = orders['order_purchase_timestamp'].dt.year
orders['order_month'] = orders['order_purchase_timestamp'].dt.month
```

```python
orders[['order_purchase_timestamp', 'order_year', 'order_month']].head()
```

```python
orders_payments = orders.merge(
    payments,
    on='order_id',
    how='left')
```

```python
orders_payments.head()
```

```python
final_df = orders_payments.merge(
    customers,
    on='customer_id',
    how='left'
)

final_df.head()
```

```python
monthly_orders = (
    final_df
    .groupby(['order_year', 'order_month'])['order_id']
    .nunique()
    .reset_index()
)

monthly_orders.head()
```

```python
monthly_orders['year_month'] = pd.to_datetime(
    monthly_orders['order_year'].astype(str) + '-' +
    monthly_orders['order_month'].astype(str) + '-01',
    format='%Y-%m-%d'
)
```

```python
monthly_orders[['order_year','order_month','year_month']].head()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(monthly_orders['year_month'], monthly_orders['order_id'])
plt.title('Monthly Orders Trend')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
final_df['order_purchase_timestamp'] = pd.to_datetime(final_df['order_purchase_timestamp'])

final_df['year_month'] = final_df['order_purchase_timestamp'].dt.to_period('M').dt.to_timestamp()
```

```python
monthly_revenue = final_df.groupby('year_month')['payment_value'].sum().reset_index()
monthly_revenue.head()
```

```python
plt.figure()
plt.plot(monthly_revenue['year_month'], monthly_revenue['payment_value'])
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
payment_revenue = final_df.groupby('payment_type')['payment_value'].sum().reset_index()
payment_revenue
```

```python
plt.figure()
plt.bar(payment_revenue['payment_type'], payment_revenue['payment_value'])
plt.title('Revenue by Payment Type')
plt.xlabel('Payment Type')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()
```

## Key Insights

### 1. Business Growth
Orders and revenue show a steady month-on-month increase, indicating healthy business expansion.

### 2. Customer Behavior
The average order value is approximately 161, suggesting frequent low-to-mid value purchases.

### 3. Payment Preferences
Credit cards contribute the highest revenue, showing strong customer preference for cashless payments.

### 4. Data Quality Handling
Missing delivery dates were associated with cancelled or undelivered orders and were handled conditionally.
